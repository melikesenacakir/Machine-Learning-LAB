```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/advertising-csv/advertising.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
print("Intercept:",mlr.intercept_)
```

```python
print("Coefficients:",list(zip(x,mlr.coef_)))
```

```python
y_pred=mlr.predict(x_test)
print("prediction for test set: {}".format(y_pred))
```

```python
mlr_dff=pd.DataFrame({"Actual values:":y_test,"Prediction values:":y_pred})
mlr_dff.head()
```

```python
from sklearn import metrics
meansqerr=metrics.mean_squared_error(y_test,y_pred)
print("R squarred:{:.2f}".format(mlr.score(x,y)*100))
print("Mean Squarred Error:",meansqerr)
```
