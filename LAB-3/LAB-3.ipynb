```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("/kaggle/input/data-csv/data.csv")
data.head()
```

```python
data.info()
```

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
```

```python
M =data[data.diagnosis=="M"]
B =data[data.diagnosis=="B"]
```

```python
M.info()
```

```python
B.info()
```

```python
plt.scatter(M.radius_mean,M.area_mean,color="red",label="malignant")
plt.scatter(B.radius_mean,B.area_mean,color="green",label="benign")
plt.legend()
plt.xlabel("radius_mean")
plt.ylabel("area_mean")
plt.show()
```

```python
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="malignant")
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="benign")
plt.legend()
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.show()
```

```python
data.diagnosis=[1 if each=="M" else 0 for each in data.diagnosis] 
```

```python
data.diagnosis
```

```python
y=data.diagnosis.values
```

```python
x_data=data.iloc[:,1:3].values
```

```python
x=(x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)) #normalize iÅŸlemi yapÄ±lÄ±yor
```

```python
from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1) #random_state=1 her seferinde aynÄ± randomlukta bÃ¶l demektir
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_head=knn.predict(x_test)
y_head
```

```python
print("when k is {}, accuracy of KNN classification {} " .format(3,knn.score(x_test,y_test))) #k deÄŸerinin knn sÄ±nÄ±flandÄ±rma doÄŸruluÄŸunu hesaplar
```

```python
test_accuracy=[]
for each in range(1,15):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    test_accuracy.append(knn2.score(x_test,y_test))

plt.figure(figsize=(5,5))
plt.plot(range(1,15),test_accuracy)
plt.title("k values vs accuracy")
plt.xlabel("k labels")
plt.ylabel("accuracy")
plt.grid()
plt.show()
print("best accuracy is {} with K ={}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
#en yÃ¼ksek doÄŸruluk deÄŸeri-> np.max(test_accuracy)
```
